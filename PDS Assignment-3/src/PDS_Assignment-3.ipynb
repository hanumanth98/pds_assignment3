**1A**

```python
import random
import pandas as pd
import matplotlib.pyplot as plt

# Set a seed for reproducibility
seed_value = 35
dframe.sample(random_state=seed_value)
# Take a random sample from the DataFrame
sample_size = 25  # Set the desired sample size
sample = dframe.sample(n=sample_size)

print(sample)

# Calculate the mean and highest Glucose values in the sample
sample_mean_glucose = sample['Glucose'].mean()
sample_max_glucose = sample['Glucose'].max()

# Display the sample statistics
print("\nSample Mean Glucose:", sample_mean_glucose)
print("Sample Highest Glucose:", sample_max_glucose)

# Calculate population statistics
population_mean_glucose = dframe['Glucose'].mean()
population_max_glucose = dframe['Glucose'].max()

# Display the population statistics
print("\nPopulation Mean Glucose:", population_mean_glucose)
print("Population Highest Glucose:", population_max_glucose)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data_comparison = {
    'Statistics': ['Mean Glucose', 'Max Glucose'] * 2,
    'Value': [population_mean_glucose, population_max_glucose, sample_mean_glucose, sample_max_glucose],
    'Group': ['Population'] * 2 + ['Sample'] * 2
}
dframe_comparison = pd.DataFrame(data_comparison)

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x='Statistics', y='Value', hue='Group', data=dframe_comparison, palette='viridis')

plt.xlabel('Statistics')
plt.ylabel('Glucose Values')
plt.title('Comparison of Population and Sample Glucose Statistics')

plt.show()
```

**1B**

```python
population_percentile_bmi = np.percentile(dframe['BMI'], 98)

sample_percentile_bmi = np.percentile(sample['BMI'], 98)

print("\nPopulation Percentile bmi:", population_percentile_bmi)
print("Sample Percentile bmi:", sample_percentile_bmi)

data_comparison = {
    'Percentile': ['98th Percentile'] * 2,
    'Value': [population_percentile_bmi, sample_percentile_bmi],
    'Group': ['Population', 'Sample']
}
dframe_comparison = pd.DataFrame(data_comparison)

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x='Percentile', y='Value', hue='Group', data=dframe_comparison, palette='viridis')
plt.xlabel('Percentile')
plt.ylabel('BMI Value')
plt.title('Comparison of 98th Percentile BMI between Population and Sample')

plt.show()
```

**1C**

```python
num_samples = 500
sample_size = 150

bootstrap_means = []
bootstrap_std_devs = []
bootstrap_percentiles = []

for _ in range(num_samples):
    bootstrap_sample = dframe['BloodPressure'].sample(n=sample_size, replace=True)
    bootstrap_means.append(bootstrap_sample.mean())

bootstrap_means = np.mean(bootstrap_means)
bootstrap_std_devs = np.std(bootstrap_means)
bootstrap_percentiles = np.percentile(bootstrap_means, 95)  # 95th percentile for example

print('mean', bootstrap_means,'sd', bootstrap_std_devs,'percentile', bootstrap_percentiles)

population_mean = dframe['BloodPressure'].mean()
population_std_dev = dframe['BloodPressure'].std()
population_percentile = np.percentile(dframe['BloodPressure'], 95)  # 95th percentile for example

print('mean', population_mean,'sd', population_std_dev,'percentile', population_percentile)

data_comparison_means = {
    'Statistic': ['Mean'] * 2,
    'Value': [bootstrap_means] + [population_mean],
    'Group': ['Bootstrap'] + ['Population']
}
dframe_comparison_means = pd.DataFrame(data_comparison_means)

data_comparison_std_devs = {
    'Statistic': ['Standard Deviation'] * 2,
    'Value': [bootstrap_std_devs] + [population_std_dev],
    'Group': ['Bootstrap'] + ['Population']
}
dframe_comparison_std_devs = pd.DataFrame(data_comparison_std_devs)

data_comparison_percentiles = {
    'Statistic': ['95th Percentile'] * 2,
    'Value': [bootstrap_percentiles] + [population_percentile],
    'Group': ['Bootstrap'] + ['Population']
}
dframe_comparison_percentiles = pd.DataFrame(data_comparison_percentiles)

sns.set(style="whitegrid")

plt.figure(figsize=(14, 8))
plt.subplot(3, 1, 1)
sns.pointplot(x='Group', y='Value', hue='Statistic', data=dframe_comparison_means, errorbar='sd', dodge=True)
plt.title('Comparison of Mean BloodPressure between Bootstrap Samples and Population')

plt.subplot(3, 1, 2)
sns.pointplot(x='Group', y='Value', hue='Statistic', data=dframe_comparison_std_devs, errorbar='sd', dodge=True)
plt.title('Comparison of Standard Deviation BloodPressure between Bootstrap Samples and Population')

plt.subplot(3, 1, 3)
sns.pointplot(x='Group', y='Value', hue='Statistic', data=dframe_comparison_percentiles, errorbar='sd', dodge=True)
plt.title('Comparison of 95th Percentile BloodPressure between Bootstrap Samples and Population')

plt.tight_layout()
plt.show()
```
